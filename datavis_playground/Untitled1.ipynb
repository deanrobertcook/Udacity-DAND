```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10)
y = x**2
y_2 = 1 / (x + 0.01)

fig, axes = plt.subplots(figsize=(8, 4), nrows=1, ncols=3, constrained_layout=True, squeeze=False)

fig.suptitle('The Ultimate Title')
axes[0][1].set_title('test')
axes[0][1].plot(y, color='Chartreuse')


vals = np.random.normal(0, 10, 10000)
axes[0][0].hist(vals, bins=30);
axes[0][1].hist(vals, bins=30);
axes[0][1].plot(y_2, color='SteelBlue')

axes[0][2].matshow(([3], [-4]));
```

```python
x = np.linspace(0, 20, 20)
y = x**2

fig = plt.figure(figsize=(6, 4))
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y, color='navy', alpha=0.5, lw=2, marker='o', markerfacecolor='y', markeredgecolor='y', markersize=12);
axes.grid(True, color='LightSteelBlue')
axes.text(0, 350, r"$\alpha$");
axes.annotate(r"$\alpha$", xy=(2.5, 50), xytext=(0, 50), arrowprops=dict(arrowstyle='->'));
#axes.set_facecolor('LightSalmon')
```
